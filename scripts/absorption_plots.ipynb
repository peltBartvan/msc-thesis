```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
def epsilon(freq, plasma_freq, scatter_freq):
    """Complex dielectric function of Drude oscillator"""
    return - plasma_freq**2/(np.square(freq) + 1j*freq*scatter_freq)
```

```python
df = pd.DataFrame(np.arange(0.5, 5, 0.02), columns = ['eV'])
df['epsilon'] = epsilon(df.eV, 2, 0.4)
df['n'] = np.real(np.sqrt(df.epsilon))
df['k'] = np.imag(np.sqrt(df.epsilon))
df.plot('eV', 'n')
```

```python
eV_range = np.linspace(0.75, 5, 100)
scatter_freq = 0.02
for plasma_freq in np.linspace(1.4, 2.2, 5):
    plt.plot(eV_range, np.imag(np.sqrt(epsilon(eV_range, plasma_freq, scatter_freq))))

plt.show()
```
